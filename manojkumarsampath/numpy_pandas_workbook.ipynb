```python
import numpy as np

# Example array
arr = np.array([[2, 3, 2], [1, 2, 4], [5, 6, 7]])
print("Array:\n", arr)

# Count even numbers
even_count = np.sum(arr % 2 == 0)
print("Even numbers count:", even_count)
```

```python
import numpy as np

# Create a 5x4 array with random integers between 1 and 100
arr = np.random.randint(1, 101, size=(5, 4))
print("2D Array:\n", arr)

col_mean = np.mean(arr, axis=0)
print("\nColumn-wise mean:", col_mean)

std_dev = np.std(arr)
print("Overall standard deviation:", std_dev)
```

```python
import pandas as pd

# Given data
data = {
    'name': ['Alice', 'Bob', 'Jefin', 'David', 'Jude'],
    'math_score': [80, 64, 90, 72, 56],
    'english_score': [78, 85, 88, 69, 60]
}

df = pd.DataFrame(data)

df.rename(columns={'name': 'student_name'}, inplace=True)

df['average'] = ((df['math_score'] + df['english_score']) / 2).astype('int8')

df.drop(columns=['math_score', 'english_score'], inplace=True)

# print final DataFrame
print(df)
```

```python
import pandas as pd

# Create the first DataFrame
data1 = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [24, 30, 22],
    'city': ['New York', 'Los Angeles', 'Chicago']
}

df1 = pd.DataFrame(data1)

# Create the second DataFrame
data2 = {
    'name': ['David', 'Eve', 'Frank'],
    'age': [28, 26, 35],
    'city': ['Houston', 'Phoenix', 'San Francisco']
}

df2 = pd.DataFrame(data2)

df_combined = pd.concat([df1, df2], ignore_index=True)

df_combined.reset_index(drop=True, inplace=True)

# print final DataFrame
print(df_combined)
```

```python
import pandas as pd

# Initialize the DataFrame with columns 'Name' and 'Age' and indices 1, 2, 3
student_info = pd.DataFrame(columns=['Name', 'Age'], index=[1, 2, 3])

# Populate the DataFrame using .loc
student_info.loc[1] = ['Alice', 24]
student_info.loc[2] = ['Bob', 30]
student_info.loc[3] = ['Charlie', 22]

student_info_sorted = student_info.sort_values(by='Age', ascending=False)

top_students = student_info_sorted.iloc[:2]['Name']
print(top_students)
```

```python
import pandas as pd

# Initialize the DataFrame with given data and custom indices
data = {
    'item': ['apple', 'banana', 'cherry', 'dates'],
    'stock': [50, 20, 0, 30],
    'price': [50, 20, 100, 150]
}

df = pd.DataFrame(data, index=['A', 'B', 'C', 'D'])

# 1) Using .loc to set the stock of 'cherry' (index "C") to 10
df.loc['C', 'stock'] = 10
print("Updated DataFrame after setting 'cherry' stock to 10:\n", df)

# 2) Using .iloc to retrieve the price value at row position 1, column position 2
price_value = df.iloc[1, 2]
print("\nPrice value at row position 1, column position 2:", price_value)
```
